```python
%pylab inline
import pandas as pd
df = pd.read_csv('responses.csv')
```

```python
X = df.as_matrix()
X = numpy.delete(X, (831), axis=0)
X.shape
```

```python
gadgets = X[:,138]
look = X[:,137]
healthy = X[:,139]
Education = df.Education
edu = zeros(1009)
cleaned = [ x for x in Education if x is not nan ]
```

```python

for i, tar in enumerate(cleaned):
    if tar[0] == 's':
        edu[i] = 0;
    if tar[0] == 'c':
        edu[i] = 1;
    if tar[0] == 'm':
        edu[i] = 2;
```

```python
size(edu)
```

```python
# plot(edu,look)
figure();
scatter(healthy, look, c=edu, alpha=0.1)
figure();
scatter(gadgets, look, c=edu, alpha=0.1)
figure();
scatter(gadgets, healthy, c=edu, alpha=0.1)
```

```python
from sklearn.model_selection import KFold
from sklearn import neighbors
X = numpy.column_stack((look,gadgets,healthy))
X = X.astype(float)
FILT = np.isnan(X)
X = X[~FILT.any(axis=1)]
C = edu[~FILT.any(axis=1)]

print (X.shape, C.shape)
```

```python
 
k_fold = KFold(n_splits=10, shuffle=True,random_state=42)
## alpha in the for loop goes through 1NN, 2NN ... 10NN. So, we can compare which one makes fewer mistake. 
for alpha in range(1,21):

    clf = neighbors.KNeighborsClassifier(alpha)

    temp = 0;

    for k, (train, test) in enumerate(k_fold.split(X)):

        Cpred = clf.fit(X[train],C[train]).predict(X[test])
 
        temp = temp + (C[test]!=Cpred).sum()
    ## compute the total number of errors in the 10 fold cv for each alpha. 
    print (alpha, 'nearest neighbor" total wrong prediction is', temp)
```

```python
##Use the Gaussian Naive Bayes classifier and compare to previous results (2 pts)
from sklearn.naive_bayes import GaussianNB


gnb = GaussianNB()



## ANOTHER WAY(MORE SAFE AND SCIENTIFIC
## temp keeps track of total number of mislabel. 
temp = 0;
## go through the K-fold cross validation.
for k, (train, test) in enumerate(k_fold.split(X)):
    ## fit(X,c) takes in the train set features and correct labels, it will compute the proper gaussian parameter 
    ## for the likelihood function, adn compute the prior as I explained above. Predict(X) takes in the test features. 
    Cpred = gnb.fit(X[train],C[train]).predict(X[test])

    temp = temp + (C[test]!=Cpred).sum()
## compute the total number of errors in the 10 cv. 
print ('total wrong prediction is', temp)


## I think the Naives Bayes and Knn are similar in performance on avaerage. 
## Some Knn happens to be better with better parameter K,  and NaiveBayes performs similar to the average 
## correct rate of KNN. 
```

```python
#Evaluate the Quadratic Discriminant Analysis the same way (2 pts)
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA


temp = 0;
for k, (train, test) in enumerate(k_fold.split(X)):

    Cpred = QDA().fit(X[train],C[train]).predict(X[test])
    temp = temp + (C[test]!=Cpred).sum()
## compute the total number of errors in the 10 cv. 
print ('total wrong prediction is', temp)

## COMPARISION
## by the observed data, it seems to have better performance than Naive Bayes, and relatively better than KNN.(depending on which KNN parameter we choose) 
```

```python
#Evaluate the Decision Tree method, too (2 pts)
from sklearn import tree
## varies the parameter about the depth of the tree. 
for depth in range(1,11):

    clf = tree.DecisionTreeClassifier(max_depth=depth)
    # store the count of mis-label.
    temp = 0
    ## perform a loop of cross validation.
    for k, (train, test) in enumerate(k_fold.split(X)):
        ## input the train set and predict on the test set. 
        Cpred = clf.fit(X[train],C[train]).predict(X[test])
        #update mislabel.
        temp = temp + (C[test]!=Cpred).sum()
    ## compute the total number of errors in the 10 cv. 
    print ('max_depth is',depth,'total wrong prediction is', temp)
    
 
```

```python
#Evaluate the Random Forest for different parameters of your choice (2 pts)
from sklearn.ensemble import RandomForestClassifier

for depth in range(1,6):
    for n_est in range(1,6):

        clf = RandomForestClassifier(n_estimators=n_est,max_depth=depth)
        temp = 0
        for k, (train, test) in enumerate(k_fold.split(X)):
            Cpred = clf.fit(X[train],C[train]).predict(X[test])
            #print (k, ':\t', (C[test]==Cpred).sum() / float(test.size))
            temp = temp + (C[test]!=Cpred).sum()
        ## compute the total number of errors in the 10 cv. 
        print ('depth =', depth, 'and n_estimatiors' ,n_est ,'total wrong prediction of is',temp)
        

```
